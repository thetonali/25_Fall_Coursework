# Elasticsearch  Search Queries , Filters & Aggregations

## Initiate modules

```python
from elasticsearch import Elasticsearch
```

## Initiate global variables

```python
client = Elasticsearch()
indexName = "medical"
docType="diseases"
searchFrom = 0
searchSize= 3
```

## Finding Lupus

### Finding Lupus first trial

```python
searchBody={
"stored_fields":["name"],
"query":{
    "simple_query_string" : {
        "query": '+fatigue+fever+"joint pain"', #疲劳 发烧 关节疼痛
        "fields": ["fulltext","title^5","name^10"]
        }
    }
}
client.search(index=indexName, body=searchBody, from_ = searchFrom, size=searchSize)
```

### Finding Lupus second trial

```python
searchBody={
#"fields":["name"],
"stored_fields":["name"],
"query":{
    "simple_query_string" : {
        "query": '+fatigue+fever+"joint pain"+rash', #皮疹
        "fields": ["fulltext","title^5","name^10"]
        }
    }
}    
#client.search(index=indexName,doc_type=docType, body=searchBody, from_ = searchFrom, size=searchSize)
client.search(index=indexName, body=searchBody, from_ = searchFrom, size=searchSize)
```

### Finding Lupus third trial

```python
searchBody={
"stored_fields":["name"],
"query":{
    "simple_query_string" : {
        "query": '+fatigue+fever+"joint pain"+rash+"chest pain"', #胸部疼痛
        "fields": ["fulltext","title^5","name^10"]
        }
    }
}
client.search(index=indexName, body=searchBody, from_ = searchFrom, size=searchSize)
```

### Fuzzy search example

```python
searchBody={
"stored_fields":["name"],
"query":{
    "simple_query_string" : {
        "query": 'lupsu~2  ',
        "fields": ["name^10"],
        }
    }
}
client.search(index=indexName,body=searchBody, from_ = searchFrom, size=searchSize)
```

## Finding Heart Attack

```python
searchBody={
"stored_fields":["name"],
"query":{
    "simple_query_string" : {
        "query": '+"left arm pain"~10',
        "fields": ["fulltext","title^5","name^10"]
        }
    },
"highlight" : {
        "fields" : {
            "fulltext" : {}
        }
    }
}
client.search(index=indexName, body=searchBody, from_ = searchFrom, size=searchSize)
```

## Finding Diabetes

```python
searchBody={
"stored_fields":["name"],
"query":{
    "simple_query_string" : {
        #"query": 'thirst "weight loss"',
        "query": 'diabetes',
        "fields": ["fulltext","title^5","name^10"]
        }
    }
}
client.search(index=indexName, body=searchBody, from_ = searchFrom, size=searchSize)
```

## Filters & Aggregations

### Diabetes KeyWords

```python
searchBody={

    "query":{
        "bool":{
            "filter": {
                     'term': {'name':'diabetes'}       

            }

        }

    },
    "aggregations" : {
        "DiseaseKeywords" : {
            "significant_terms" : { "field" : "fulltext", "size" : 30 }
        }
    }
}



client.search(index=indexName, body=searchBody, from_ = searchFrom, size=searchSize)
```

return here after creating the diseases2 doctype

```python
client = Elasticsearch()
indexName = "medical"
#docType="diseases2"  新版基本无用
searchFrom = 0
searchSize= 3
```

```python
searchBody={
 "query":{
        "bool":{
            "filter": {
                     'term': {'name':'diabetes'}       

            }

        }

    }, 
"aggregations" : {
        "DiseaseKeywords" : {
            "significant_terms" : { "field" : "fulltext", "size" : 30 }
        },
        "DiseaseBigrams": {
            "significant_terms" : { "field" : "fulltext.shingles", "size" : 30 }
        }
    }
}
client.search(index=indexName, body=searchBody, from_ = searchFrom, size=searchSize)
```
