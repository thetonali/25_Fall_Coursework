```python
# python三种输出格式
# 1.字符串格式化(f-string)
# 在字符串前加上字母 f，然后在花括号 {} 中直接包含变量名。
name='张三'
age=33
print(f'我的名字是{name},今年{age}岁了。')
print(f"{age+123000:,}")
print(f"{age:.2f}")
print(f"{'你好'}{name:^10}")
```

```python
# 2.str.format()方法
# 在字符串中使用花括号 {} 作为占位符，
# 并在字符串后调用 .format() 方法并传入变量。
name='李四'
age=44
print("{}的年龄是{}岁。".format(name,age))
print("{1}的年龄是{0}岁。".format(age,name))
print("{n}的年龄是{a}岁。".format(n=name,a=age))

# "{:.2f}"保留两位小数
```

```python
# 3.百分号%操作符
# 使用类似C语言的格式说明符，如 %s (字符串)、%d (整数)、%f (浮点数) 等，
# 并在字符串后使用 % 运算符加变量（或变量元组）。
name='王五'
age=55
print("我的名字是%s,今年%d岁了。"%(name,age))
```
