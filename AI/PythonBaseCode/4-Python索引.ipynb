### 索引
* 从前面是0开始
* 从后面是-1开始

```python
tang = 'tang yu di'
tang[0]
# 正数第i+1个，倒数第i个
```

```python
tang[5]
```

```python
tang[-1]
```

```python
tang[-5]
```

```python
tang[10]
# 超出范围
```

### 切片
：表示从哪到哪
左闭右开的区间

```python
tang='Charles Leclerc'
tang[0:4]
# 左闭右开
```

```python
tang[5:]
```

```python
tang[:7]
# 依然左闭右开
```

```python
tang[1:-2]
# 左闭右开
```

```python
tang[-3:]
```

```python
tang[:]
```

```python
tang[::2]
```

```python
tang[1:9:3]
tang[1:9]
```
